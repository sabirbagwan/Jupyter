# 4. Automating with Selenium


```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

website = 'https://URL of the website'
path = r'C:\Users\Downloads\chromedriver'

service = Service(executable_path = path)
driver = webdriver.Chrome(service = service)

driver.get(website)

username = driver.find_element(By.ID, "name")
password = driver.find_element(By.ID, "password")

username.send_keys("YourUserName")
password.send_keys("Password")

driver.find_element(By.NAME, "submit").click()
```

# 5. Making the pivot table from a xlsx file and exporting it to another one

```python
import pandas as pd

df = pd.read_excel('filename.xlsx')
df = df[['Gender', 'Product', 'Total']]

pivot_table = df.pivot_table(index = 'Gender', columns = 'Product', values = 'Total', aggfunc = 'sum').round(0)

pivot_table.to_excel('pivot_table.xlsx', 'Report_name', startrows = 4)
```
